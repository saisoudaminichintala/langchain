### Streaming

- Most models can stream the output content while it is being generated. 
- By displaying output progressively, streaming significantly imroves user experience , particularly for longer responseds.
- Calling Stream() returns an iterator that yields output chunks as they are produced . you can use a loop to process each chunk in real time

```python
import os
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage

load_dotenv()
os.environ["OPENAI_API_KEY"]=os.getenv("OPENAI_API_KEY")
```

```python
model = ChatOpenAI(
    model= "gpt-5"

)

for chunk in model.stream("why do parrots have colors"):
    print(chunk.content, end="|" , flush=True)
```

### Batch

- Batching a collection of independent requeata  to a model can significantly improve preformance, reduce costs  as the processign can be done in Parallel

```python
responses = model.batch([
    "why do parrots have colors?",
    " what is quantum computing?",
    "How do Aeroplanes fly?"],
    config={
        'max_concurrency':5 , # Limit to 5 parallel calls
    }
)

for response in responses:
    print(response)
```
